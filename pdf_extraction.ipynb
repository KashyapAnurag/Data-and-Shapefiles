```python
import os
import pdfplumber
import pandas as pd

base_dir = r"C:\Users\INAK09124\OneDrive - WSP O365\PERSONAL\Exports\Creagh Manholes Issue 1\Assets\Manholes"
output_file = r"C:\Users\INAK09124\OneDrive - WSP O365\PERSONAL\Exports\ISSUE1_try2.xlsx"

data = []

for folder in os.listdir(base_dir):
    folder_path = os.path.join(base_dir, folder)
    
    if os.path.isdir(folder_path):
        for pdf_file in os.listdir(folder_path):
            if pdf_file.endswith('.pdf'):
                pdf_path = os.path.join(folder_path, pdf_file)
                
                with pdfplumber.open(pdf_path) as pdf:
                    if len(pdf.pages) > 0:
                        page_1_text = pdf.pages[0].extract_text()
                        lines = page_1_text.split('\n')
                        file_data = [pdf_file] + lines
                        data.append(file_data)

columns = ['File Name'] + [f'Line {i+1}' for i in range(max(len(d) - 1 for d in data))]
df = pd.DataFrame(data, columns=columns)
df.to_excel(output_file, index=False)

print(f"Data extraction complete. Excel file saved as '{output_file}'")
```
